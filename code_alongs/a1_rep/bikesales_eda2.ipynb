#Profit per age-group

```python
import duckdb

df_small_groups = duckdb.query("""
    SELECT
        age_group,
        customer_gender, 
        country,
        SUM(profit) AS Total_Profit,
        COUNT(*) AS Number_Rows
    FROM df
    GROUP BY
        age_group, customer_gender, country
    -- HAVING används för filter på aggregerade värden
    HAVING
        COUNT(*) < 1000
    ORDER BY
        Total_Profit DESC
""").df()

df_small_groups
```

```python
df_senior_female = df_small_groups.query("Age_Group =='Seniors (64+)' and Customer_Gender == 'F'")

df_senior_female
```

```python
ax = df_senior_female.plot(kind = 'barh', x = "Country", y = "Total_Profit") #.plot = wrapper runt matplotlib
ax.set(
    title = "64 + females total profit in bike store",
    xlabel = "Total profit",
    ylabel = "Country",
)

ax.invert_yaxis()
fig= ax.get_figure()
fig.savefig("profit_per_country_females.png", dpi = 200)
```
