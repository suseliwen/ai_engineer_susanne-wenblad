  b) Draw a line chart of quantity for each year. Is there a year that should be skipped?



```python
months_per_year = df.groupby("Year")["Month"].nunique()
months_per_year
```

```python
yearly = (df
          .groupby("Year", as_index = False) ["Quantity"]      # ("Year = skapar en grupp per år/ ["Quantity = summerar antalet inom varje år"]")
                                                               # as_index --> gör att Year ligger kvar som vablig kolumn och inte indec
          .sum()
)                                                              # Resultat: yearly har två kolumner: Year och Quantity (den summerade).

yearly_cleaned = yearly[yearly["Year"] != 2017]

```

```python

fig, ax = plt.subplots()
ax.plot(yearly_cleaned["Year"], yearly_cleaned["Quantity"], marker = "o")
ax.set(
    title = "Nya bilregistreringar per år (Norge)",
    xlabel = "År",
    ylabel = "Registreringar"
)

plt.show
```

  c) Draw a line chart of average CO2 emissions for same years that as in b)



```python
avg_co = (df
          .groupby("Year", as_index= False) ["Avg_CO2"]
          .sum()

)
avg_co = avg_co[avg_co["Year"] != 2017]

avg_co
```

```python
fig, ax = plt.subplots()
ax.plot(avg_co["Year"], avg_co["Avg_CO2"], marker = "o")
ax.set(
    title = "Utsläpp per år",
    xlabel = "År",
    ylabel = "Utsläpp"
)

plt.show()
```

  d) Draw a line chart of all years and months for import



```python

df = df[df["Year"] != 2017]

monthly = (df.groupby(["Year", "Month"], as_index=False)["Import"].sum())

pivot = monthly.pivot(index="Month", columns="Year", values="Import")

pivot
```

```python
fig, ax = plt.subplots()
for year in pivot.columns:
    ax.plot(pivot.index, pivot[year], marker = "o", label=str(year))

ax.set(title = "Import per månad", xlabel="Månad", ylabel="Import")
ax.legend(title= "År")
plt.show()
```

  e) Draw a line chart of all years and months for average CO2 emissions



```python

avg_co2 = (df.groupby(["Year", "Month"], as_index=False)["Avg_CO2"].sum())

pivot = avg_co2.pivot(index= "Month", columns= "Year", values= "Avg_CO2")

pivot
```

```python
fig, ax = plt.subplots()

for year in pivot.columns:
    ax.plot(pivot.index, pivot[year], marker = "o", label = str(year))

ax.set(title = "Utsläpp av co2 per år", xlabel = "månad", ylabel ="co2")
ax.legend(title= "År")
plt.show()
```

  f) Draw a line chart of all years and months for electric cars import where it's relevant.



```python
df_e = df[df["Import_Electric"].notna()]

electric_import = (df_e.groupby(["Year", "Month"], as_index=False)["Import_Electric"].sum())

pivot = electric_import.pivot(index = "Month", columns = "Year", values= "Import_Electric")

pivot = pivot.fillna(0)

pivot

```

```python
fig, ax = plt.subplots()

for year in pivot.columns:
    ax.plot(pivot.index, pivot[year], marker = "o", label= str(year))

ax.set(title = "Import av elbilar per år", xlabel = "månad", ylabel="Import_Electric")
ax.legend(title = "År")
plt.show()
```

  g) Draw a line chart of average diesel share per year



```python
#Diesel_Share

mean_diesel = (df.groupby(["Year"], as_index= False)["Diesel_Share"].sum())

mean_diesel

```

```python
fig, ax = plt.subplots()
ax.plot(mean_diesel["Year"], mean_diesel["Diesel_Share"], marker = "o")
ax.set(
    title = "Diesel_Share",
    xlabel = "År",
    ylabel = "Diesel_Share"
)

plt.show()

```
